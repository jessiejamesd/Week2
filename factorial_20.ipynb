```python
""" function to calculate the factorial of a number (a nonnegative integer) """
def factorial(n):
  f=1
  for x in range(1,n):
    f*=x
  return f
num = int(input("Enter a number:"))
if(num>0):
  print("The factorial of",num,"is:",factorial(num+1))
else:
  print("Negative numbers are not allowed!")
```
