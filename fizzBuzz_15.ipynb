```python
""" program which iterates from 1 to 10. For multiples of 2, print
“Fizz” instead of the number and for the multiples of 5, print “Buzz”. For
numbers which are multiples of both 2 and 5, print “FizzBuzz”. """
for x in range(1,11):
  if(x%2==0 and x%5==0):
    print("FizzBuzz",end=" ")
  elif(x%2==0):
    print("Fizz",end=" ")
  elif(x%5==0):
    print("Buzz",end=" ")
  else:
    print(x,end=" ")
```
